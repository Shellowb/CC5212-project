```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("DATA/q1-std-in-temperature-by-country.tsv", sep='\t', header=None, engine='python', encoding='utf-8', error_bad_lines=False) 
df.columns = ["country", "sig_var"]

print('Percentage of missing data')
print((df.isnull().sum() / len(df))*100)
df
```

```python
df.sort_values(by=['sig_var'], ascending=False, inplace=True)
```

```python
places = np.unique(df["country"])
print(places.size)
places
```

```python
top = df.sort_values(by=['sig_var'], ascending=False).head(30)["country"]
bot = df.sort_values(by=['sig_var'], ascending=True).head(30)["country"]
tips = list(top) + list(bot)
top
```

```python
df[df["country"].isin(top)]
```

```python
#sns.set(rc={'figure.figsize':(11.7,8.27)})
f, ax = plt.subplots(figsize=(8, 12))
sns.barplot(x="sig_var", y="country", data=df[df["country"].isin(top)], palette=sns.color_palette("coolwarm",30), ax=ax)
```
